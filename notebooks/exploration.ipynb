```python
import pathlib
from pypdf import PdfReader
import re
```

#### Encontrar archivos pdfs en una carpeta

```python
folder_path: str = '../data'
pdfs = list(pathlib.Path(folder_path).glob("*.pdf"))
pdfs
```

#### Extraer el texto de un archivo pdf

```python

reader = PdfReader(pdfs[0])

texto_completo = ""
for pagina in reader.pages:
    texto_completo += pagina.extract_text() + "\n"

print(texto_completo)
```

#### Extraer texto con regex

```python
match = re.search(r'INVOICE\s*\n\s*#\s*(\d+)', texto_completo, re.DOTALL | re.MULTILINE)

print(match.group(1) if match else "No encontrado")
```
